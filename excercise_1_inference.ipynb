```python
import torch
```

```python
from excercise_1_seq2seq import Decoder, Encoder, Seq2SeqManager, prepare_model_input, SOS_IDX, outputs2words, VOCAB_SIZE
```

```python
def predict_word(model, word, device, max_len=20):
    model.eval()
    with torch.no_grad():
        src_tensor = prepare_model_input([word]).to(device)

        dummy_trg = torch.zeros((1, max_len), dtype=torch.long).to(device)
        dummy_trg[:, 0] = SOS_IDX

        # Shape: [1, max_len, vocab_size]
        output = model(src_tensor, dummy_trg, teacher_forcing=False)

        # Greedy prediction, Shape: [1, max_len]
        prediction_indices = output.argmax(2)

        predicted_word = outputs2words(prediction_indices)[0]

        return predicted_word
```

```python
EMB_DIM = 64
HID_DIM = 64
N_LAYERS = 2
```

```python
encoder = Encoder(VOCAB_SIZE, EMB_DIM, HID_DIM, N_LAYERS)
decoder = Decoder(VOCAB_SIZE, EMB_DIM, HID_DIM, N_LAYERS)
model = Seq2SeqManager(encoder, decoder, VOCAB_SIZE)
```

```python
save_path = 'weights/corrupted_word_model.pth'
model.load_state_dict(torch.load(save_path))
```

```python
while True:
    user_input = input('type "q" to quit, input word: ')
    if user_input == 'q' or not user_input:
        break

    correction = predict_word(model, user_input, device='cpu')
    print(f"Input: {user_input} -> Prediction: {correction}")
```
